```python
import numpy as np
import scipy as sp
import math
```

```python
#For Validation
Archaludon = [90,105,130,125,65,85,'Archaludon']
Enteisaur = [105,115,121,65,93,91,'Gouging Fire']
Raikousaur = [125,73,91,137,89,75,'Raging Bolt']
RoboTerra = [90,120,80,68,108,124,'Iron Boulder']
RoboCobal = [90,72,100,122,108,98,'Iron Crown']
Terapagos = [160,105,110,130,110,85,'Terapagos']
```

```python
#PoGo Conversion
def pogostats(pokemon):
    Attack = [pokemon[1], pokemon[3]]
    Defense = [pokemon[2], pokemon[4]]
    HP = pokemon[0]
    Spe = pokemon[5]
    Name = pokemon[6]
    def stat_conversion(Attack,Defense,HP,Spe):
        ScaledAttack = round(2*((7/8*max(Attack) + 1/8*min(Attack))))
        SpeedMod = 1+((Spe-75)/500) 
        BaseAttack = round(ScaledAttack * SpeedMod)
        ScaledDefense = round(2*((5/8*max(Defense) + 3/8*min(Defense))))
        BaseDefense = round(ScaledDefense * SpeedMod)
        BaseStamina = np.floor(HP*1.75 + 50)
      #CP Determination
        CPM =0.840300010 #at lvl50
        CP_Atk = (BaseAttack + 15)
        CP_Def = (BaseDefense + 15)
        CP_HP = (BaseStamina + 15)
        CP = np.floor((((CP_HP**0.5)*(CP_Def**0.5)*CP_Atk*CPM**2))/10)
        CP_40 = np.floor((((CP_HP**0.5)*(CP_Def**0.5)*CP_Atk*(0.7903**2)))/10)
    
        if CP_40 >= 4000:
            BaseAttack = round(BaseAttack*0.91)
            BaseDefense = round(BaseDefense*0.91)
            BaseStamina = round(BaseStamina*0.91)
            CP_Atk = (BaseAttack + 15)
            CP_Def = (BaseDefense + 15)
            CP_HP = (BaseStamina + 15)
            CP = np.floor((((CP_HP**0.5)*(CP_Def**0.5)*CP_Atk*CPM**2))/10)
         
        return BaseAttack, BaseDefense, int(BaseStamina) , int(CP)
    
    def interpolate_value(value, min_value, max_value):
    # Ensure the value is within the range [min_value, max_value]
    
        value = max(min(value, max_value), min_value)
        
        # Linear interpolation formula
        percentage = (value - min_value) / (max_value - min_value)
        
        return percentage

    
    result = stat_conversion(Attack,Defense,HP,Spe)
    attack_benchmark = round(interpolate_value(result[0],17,345),2) 
    defense_benchmark = round(interpolate_value(result[1],32,396),2) 
    hp_benchmark = round(interpolate_value(result[2],67,496),2) 
    cp_benchmark = round(interpolate_value(result[3],310,5010),2)
    
    print (Name)
    print ('Attack:',result[0] )
    print ('Defense:',result[1] )
    print ('HP:',result[2] )
    print ('Max CP:', result[3])
    print(f"Attack of {result[0]} is located at {attack_benchmark * 100}% within the distribution.")
    print(f"Defense of {result[1]} is located at {defense_benchmark * 100}% within the distribution.")
    print(f"HP of {result[2]} is located at {hp_benchmark * 100}% within the distribution.")
    print(f"Max CP of {result[3]} is located at {cp_benchmark * 100}% within the distribution.")      
```

```python
#Testing
print(pogostats(Archaludon))
print(pogostats(Enteisaur))
print(pogostats(Raikousaur))
print(pogostats(RoboTerra))
print(pogostats(RoboCobal))
print(pogostats(Terapagos))
```
